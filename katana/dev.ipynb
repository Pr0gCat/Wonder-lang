```python
import lark
```

```python
parser = lark.Lark(open('../larks/preprocess.lark', 'r').read(), start='_start', import_paths=['../larks'])
tree = parser.parse("""
        @md
        @error(\"abc\")
        asd
        abc = @exec(\"ls\")
        asd @lib
        @if 
        \"\"\"\"\"\"
        @end
        @md
    @end
    \"\"\"
        LOL
    \"\"\"
    """)
print(tree)
```

```python
def preprocess(fp, tree):
    for node in tree.children:
        t = type(node) 
        if t is lark.Tree:
            fp.write(f'<ss_func - {node.children}>')
            continue
        # is token
        if node.type == 'CODE' or node.type == 'NL':
            fp.write(node.value)
        else: 
            fp.write(f'<{node.value}>')
import io
f = io.StringIO()
preprocess(f, tree)
print(f.getvalue())
```
